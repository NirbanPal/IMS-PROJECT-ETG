```python
fd=open("frecord.json","r")
txt=fd.read()
fd.close()
```

```python
import json
records=json.loads(txt)
```

```python
records
```

```python
ps=int(input("how many products you wanna add:"))
for g in range (0,ps):
  pd_id=str(input("enter your product id:"))
  name=str(input("enter product name:"))
  pd_co=str(input("enter product company:"))
  pr=int(input("enter the price:"))
  for m in records.keys():
    if pd_id == m :
      paq=records[pd_id]["quantity"]
  qn=int(input("enter quantity:"))
  fas=str(input("fassi certified:"))
  for n in list(records):
    if pd_id == n :
      records[pd_id]={"name":name,"product company":pd_co,"price":pr,"quantity":qn+paq,"fassi certified":fas}
    else:
      records[pd_id]={"name":name,"product company":pd_co,"price":pr,"quantity":qn,"fassi certified":fas}
```

```python
import json
js=json.dumps(records)
```

```python
fd=open("frecord.json","w")
fd.write(js)
fd.close()
```
